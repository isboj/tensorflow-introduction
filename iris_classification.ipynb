インポート

```python
import urllib

import tensorflow as tf
import numpy as np
import pandas as pd
```

データ取得用URLの定義

```python
IRIS_TRAINING = "iris_training.csv"
IRIS_TRAINING_URL = "http://download.tensorflow.org/data/iris_training.csv"

IRIS_TEST = "iris_test.csv"
IRIS_TEST_URL = "http://download.tensorflow.org/data/iris_test.csv"
```

データ取得および保存(訓練用データ)

```python
raw = urllib.request.urlopen(IRIS_TRAINING_URL).read()
with open(IRIS_TRAINING,'wb') as f:
    f.write(raw)
```

データ取得および保存(テスト用データ)

```python
raw = urllib.request.urlopen(IRIS_TEST_URL).read()
with open(IRIS_TEST,'wb') as f:
    f.write(raw)
```

保存したファイルをデータフレーム型で読み込む(Pandas利用)

訓練データ

```python
iris_train = pd.read_csv(IRIS_TRAINING)
```

```python
iris_train
```

テストデータ

```python
iris_test = pd.read_csv(IRIS_TEST)
```

```python
iris_test
```

訓練データから正解ラベルを分ける

```python
iris_train_x = iris_train.iloc[:,:4]
```

訓練データの正解ラベル(ダミー変数化する)

```python
iris_train_t = iris_train.iloc[:,4:]
iris_train_t = pd.get_dummies(iris_train_t["virginica"])
```

テストデータから正解ラベルを分ける

```python
iris_test_x = iris_test.iloc[:,:4]
```

テストデータの正解ラベル(ダミー変数化する)

```python
iris_test_t = iris_test.iloc[:,4:]
iris_test_t = pd.get_dummies(iris_test_t["virginica"])
```

##ここから、分類アルゴリズムの定義

確率の計算式の定義

```python
x = tf.placeholder(tf.float32, [None, 4])
w = tf.Variable(tf.zeros([4, 3]))
w0 = tf.Variable(tf.zeros([3]))
f = tf.matmul(x, w) + w0
p = tf.nn.softmax(f)
```

誤差関数と最適化関数の定義

```python
t = tf.placeholder(tf.float32, [None, 3])
loss = -tf.reduce_sum(t*tf.log(p))
train_step = tf.train.AdamOptimizer().minimize(loss)
```

正解率の算出を定義

```python
correct_prediction = tf.equal(tf.argmax(p, 1), tf.argmax(t,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
```

初期化をしたのちに、トレーニングを開始
一定ステップごとに誤差と正解率を表示

```python
sess = tf.InteractiveSession()
sess.run(tf.global_variables_initializer())
i = 0
for _ in range(2000):
    i += 1
    sess.run(train_step, feed_dict={x: iris_train_x , t: iris_train_t})
    if i % 100 == 0:
        loss_val, acc_val = sess.run([loss, accuracy],
                                    feed_dict = {x: iris_test_x, t: iris_test_t})
        print("Step: %d, Loss: %f, Accuracy: %f" %(i, loss_val, acc_val))
```

実際にどのように予測されているのかの確認

```python
sess.run(tf.argmax(p, 1), feed_dict = {x: iris_test_x})
```
